```python
import pandas as pd #importing pandas

df = pd.read_csv(r"C:\Users\admin\Desktop\Finance Project\finance_project_dataset.csv")
df.head() #first five entries
```

```python
#Inspect the structure of the dataset

df.shape #rows and columns in the dataset
df.columns #column names of dataframes
df.info() #summary of dataset structure
df.describe() #statistical summary for numerical columns - mean, count, avg, etc
```

```python
df.isnull().sum() #chk for null values
df.drop_duplicates(inplace=True) #drop the duplicates from the datasets if exist
```

```python
df['Date'] = pd.to_datetime(df['Date']) #change the type of order_date to date instead of object
df.info()
```

```python
df['Month'] = df['Date'].dt.month #create a new column Month by extracting it
df.columns #column names of dataframes
```

```python
df.to_csv("cleaned_finance_data.csv", index=False) #save the cleaned file
```

```python
#EDA Analysis
df['Amount'].sum() #total transaction value
```

```python
df.groupby('Account_Type')['Amount'].sum() #sum of transactions by account type
```

```python
df.groupby('Branch')['Amount'].sum() #sum of transactions by branch
```

```python
df.groupby('Month')['Amount'].sum() #monthly trend of transactions
```

```python
df.groupby('Customer_ID')['Amount'].sum().sort_values(ascending=False).head(10) #top ten high value customers by transaction amount
```
